```python
def CategoryTemp(x):
    if (x <= 20):
        return ('A -2 - 20')
    elif (x > 20) & (x <= 40):
        return ('B 20 - 40')
    elif (x > 40) & (x <= 60):
        return ('C 40 - 60')
    elif (x > 60) & (x <= 80):
        return ('D 60 - 80')
    elif (x > 80) & (x <= 101):
        return ('E 80-101')
```

```python
df['CategoryTemp'] = df['Temperature'].apply(CategoryTemp)
df.head()
```

```python
def CategoryFuel(x):
    if (x <= 3):
        return ('A 2.4 - 3')
    elif (x > 3) & (x <= 3.5):
        return ('B 3 - 3.5')
    elif (x > 3.5) & (x <= 4):
        return ('C 3.5 - 4')
    elif (x > 4) & (x <= 4.5):
        return ('D 4 - 4.5')
```

```python
df['CategoryFuel'] = df['Fuel_Price'].apply(CategoryFuel)
df.head()
```

```python
def CategoryCPI(x):
    if (x <= 140):
        return ('A 120 - 140')
    elif (x > 140) & (x <= 160):
        return ('B 140 - 160')
    elif (x > 160) & (x <= 180):
        return ('C 160 - 180')
    elif (x > 180) & (x <= 200):
        return ('D 180 - 200')
    elif (x > 200) & (x <= 220):
        return ('E 200 - 220')
    elif (x > 220) & (x <= 240):
        return ('F 220 - 240')
```

```python
df['CategoryCPI'] = df['CPI'].apply(CategoryCPI)
df.head()
```

```python
def CategoryUnemp(x):
    if (x <= 6):
        return ('A 4 - 6')
    elif (x > 6) & (x <= 8):
        return ('B 6 - 8')
    elif (x > 8) & (x <= 10):
        return ('C 8 - 10')
    elif (x > 10) & (x <= 12):
        return ('D 10 - 12')
    elif (x > 12) & (x <= 14):
        return ('E 12 - 14')
    elif (x > 14) & (x <= 16):
        return ('F 14 - 16')
```

```python
df['CategoryUnemp'] = df['Unemployment'].apply(CategoryUnemp)
df.head()
```
